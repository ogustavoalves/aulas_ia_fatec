## Data visualization

### Distribuição de variável "Company Name"

```python
plt.figure(figsize=(12, 6))  # Define um tamanho adequado antes de criar o gráfico

fig = sns.histplot(data=df, x='Company Name', kde=True, color='b', bins=30)  # Ajuste o número de bins se necessário

fig.set_title('Distribuição da variável', size=18)
fig.set_xlabel('Screen Size', size=14)
fig.set_ylabel('Count', size=14)
plt.xticks(rotation=90)  

plt.tight_layout()  
plt.show()
```

### Distribuição de variável

```python
plt.figure(figsize=(10, 36))
fig = sns.histplot(data=df, y = 'Processor', color='b')
fig.set_title('Distribuição da variável "Processor"', size='18')

fig.set_xlabel('Count', size=14)
fig.set_ylabel('Processor', size=14)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9, 5))
fig = sns.histplot(data=df, x="Launched Year", color='g', kde=True)
fig.set_title('Distribuição da variável "Year Lauched"', size=18)

fig.set_xlabel('Year')
fig.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

## Método `select_dtypes`

```python
df_numerical = df.select_dtypes(exclude = 'object').columns

for i in df_numerical:
    fig = sns.histplot(data=df, x=i)
```
